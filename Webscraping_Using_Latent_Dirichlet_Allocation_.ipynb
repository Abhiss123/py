```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from gensim import corpora, models
from gensim.parsing.preprocessing import remove_stopwords, strip_punctuation

# Download NLTK data
nltk.download('stopwords')
nltk.download('punkt')
```

```python
url = 'https://thatware.io/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
#soup.find_all('p') is like a magnet that attracts all the paragraph elements from the HTML content, and stores them in a list for later use!

paragraphs = soup.find_all('p')
paragraphs
```

```python
#[para.get_text() for para in paragraphs]: This creates a list of all the extracted texts. For example, if the paragraphs list contains three paragraph elements with the texts.
#"Hello World", "This is a test", and "SEO framework", this step would create the list ["Hello World", "This is a test", "SEO framework"].

#' '.join(...): This joins all the texts in the list into a single string, separated by spaces.
#Using the previous example, this step would create the string "Hello World This is a test SEO framework".

text = ' '.join([para.get_text() for para in paragraphs])
text
```

```python
# Converts the whole text into lower case.

text = text.lower()
text
```

```python
#strip_punctuation is a function that removes punctuation marks from the text. Punctuation marks are symbols like ., ,, !, ?, ;, etc.

text = strip_punctuation(text)
text
```

```python
#remove_stopwords(text) is a function that removes common words like "the", "and", "a", "is", etc. from the text.
#These words are called "stop words" because they don't add much meaning to the text.

text = remove_stopwords(text)
text
```

```python
#nltk.word_tokenize(text) is a function that breaks down a piece of text into individual words or tokens.
#Input text: "Hello World, this is a test."
#Output tokens: ["Hello", "World", ",", "this", "is", "a", "test", "."]

tokens = nltk.word_tokenize(text)
tokens
```

```python
def preprocess_text(text):
    text = text.lower()
    text = strip_punctuation(text)
    text = remove_stopwords(text)
    tokens = nltk.word_tokenize(text)
    return tokens
```

```python
#Think of it like a word collector that gathers all the unique words from the text and stores them in a dictionary
#Input tokens: ["Hello", "World", "this", "is", "a", "test", "Hello", "World"]
#Output dictionary: {"Hello": 0, "World": 1, "this": 2, "is": 3, "a": 4, "test": 5}

dictionary = corpora.Dictionary([tokens])
dictionary
```

###**Corpus = [dictionary.doc2bow(tokens)]** is a step in natural language processing (NLP) that prepares text data for analysis

**Doc2bow:** This stands for "document to bag of words". It's a way to represent a document (like a sentence or text) as a collection of words, without considering the order of the words.

**Dictionary:** This is the vocabulary we created earlier, which contains all the unique words from the text.

**Tokens:** These are the individual words extracted from the text.

**dictionary.doc2bow(tokens)**, it creates a "bag of words" representation of the text. This means it counts the number of times each word appears in the text, without considering the order of the words.

Input tokens: ["Hello", "World", "this", "is", "a", "test", "Hello", "World"]

Output corpus: [(0, 2), (1, 2), (2, 1), (3, 1), (4, 1), (5, 1)]


- "Hello" (index 0) appears twice
- "World" (index 1) appears twice
- "this" (index 2) appears once
- And so on...

```python
def find_lda_keyword_relevance(text, keyword, num_topics=5):
    tokens = preprocess_text(text)
    dictionary = corpora.Dictionary([tokens])
    corpus = [dictionary.doc2bow(tokens)]

    lda = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
    topics = lda.print_topics(num_words=10)

    keyword_relevance = {}
    for idx, topic in topics:
        words = topic.split(' + ')
        for word in words:
            word = word.split('*')[1].strip('"')
            if word in keyword.split():
                if word in keyword_relevance:
                    keyword_relevance[word] += 1
                else:
                    keyword_relevance[word] = 1

    return keyword_relevance
```

**lda = models.LdaModel(corpus, num_topics=50, id2word=dictionary, passes=100)** is a step in natural language processing (NLP) that performs topic modeling using Latent Dirichlet Allocation (LDA).

**LdaModel:** This is a class from the Gensim library that implements the LDA algorithm.

**corpus:** This is the text data that we processed and converted into a format that the LDA model can understand (bag-of-words).

**num_topics=5:** This tells the LDA model how many topics we want it to find in our text data. For example, if we set num_topics=5, it will try to find 5 different topics.

**id2word=dictionary:** This is the dictionary we created, which maps each word to a unique ID. The LDA model uses this dictionary to understand the words in our corpus.

**passes=100:** This tells the LDA model how many times it should go through the entire corpus to refine its understanding of the topics. More passes can lead to better results but take more time.

**Why is this step used?**

The LDA model helps us find hidden topics in a large collection of text. It's like a tool that reads through a lot of text and identifies groups of words that frequently appear together, which represent different topics.

**topics = lda.print_topics(num_words=10)**

**lda.print_topics(num_words=10):** This function tells the LDA model to print the top 10 words for each topic it found. These words are the ones that best represent each topic.

**Why is this step used?**

This step helps us see the actual topics that the LDA model discovered. By looking at the top words for each topic, we can understand what each topic is about.

```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from gensim import corpora, models
from gensim.parsing.preprocessing import remove_stopwords, strip_punctuation
from gensim.models.coherencemodel import CoherenceModel
from collections import defaultdict

# Download NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Function to scrape website content
def scrape_website(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')
    text = ' '.join([para.get_text() for para in paragraphs])
    return text

# Function to preprocess text
def preprocess_text(text):
    text = text.lower()
    text = strip_punctuation(text)
    text = remove_stopwords(text)
    tokens = nltk.word_tokenize(text)
    return tokens

# Function to apply LDA and find keyword relevance
def find_lda_keyword_relevance(text, keyword, num_topics=100):
    tokens = preprocess_text(text)
    dictionary = corpora.Dictionary([tokens])
    corpus = [dictionary.doc2bow(tokens)]

    lda = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
    topics = lda.print_topics(num_words=247)

    # Print tokens and topics for debugging
    print("Tokens:", tokens)
    print("Topics:")
    for topic in topics:
        print(topic)

    keyword_relevance = defaultdict(float)
    keyword_tokens = keyword.lower().split()
    for idx, topic in topics:
        words = topic.split(' + ')
        for word in words:
            prob, term = word.split('*')
            term = term.strip('"')
            if term in keyword_tokens:
                keyword_relevance[term] += float(prob)

    return keyword_relevance, lda, dictionary, tokens

# Function to calculate coherence score
def calculate_coherence_score(lda_model, texts, dictionary, coherence='c_v'):
    coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=dictionary, coherence=coherence)
    coherence_lda = coherence_model_lda.get_coherence()
    return coherence_lda

# URL and keyword
url = 'https://timesofindia.indiatimes.com/'
keyword = 'News'

# Scrape website content
text = scrape_website(url)

# Preprocess and find LDA relevance
keyword_relevance, lda_model, dictionary, texts = find_lda_keyword_relevance(text, keyword)

# Calculate coherence score
coherence_score = calculate_coherence_score(lda_model, [texts], dictionary)

# Output results
print(f'Keyword relevance in LDA topics for "{keyword}": {keyword_relevance}')
print(f'Coherence Score for LDA model: {coherence_score}')
```

```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from gensim import corpora, models
from gensim.parsing.preprocessing import remove_stopwords, strip_punctuation
from gensim.models.coherencemodel import CoherenceModel
from collections import defaultdict

# Download NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Function to scrape website content
def scrape_website(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')
    text = ' '.join([para.get_text() for para in paragraphs])
    return text

# Function to preprocess text
def preprocess_text(text):
    text = text.lower()
    text = strip_punctuation(text)
    text = remove_stopwords(text)
    tokens = nltk.word_tokenize(text)
    return tokens

# Function to apply LDA and find keyword relevance
def find_lda_keyword_relevance(text, keyword, num_topics=100):
    tokens = preprocess_text(text)
    dictionary = corpora.Dictionary([tokens])
    corpus = [dictionary.doc2bow(tokens)]

    lda = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
    topics = lda.print_topics(num_words=247)

    # Print tokens and topics for debugging
    print("Tokens:", tokens)
    print("Topics:")
    for topic in topics:
        print(topic)

    keyword_relevance = defaultdict(float)
    keyword_tokens = keyword.lower().split()
    for idx, topic in topics:
        words = topic.split(' + ')
        for word in words:
            prob, term = word.split('*')
            term = term.strip('"')
            if term in keyword_tokens:
                keyword_relevance[term] += float(prob)

    return keyword_relevance, lda, dictionary, tokens

# Function to calculate coherence score
def calculate_coherence_score(lda_model, texts, dictionary, coherence='c_v'):
    coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=dictionary, coherence=coherence)
    coherence_lda = coherence_model_lda.get_coherence()
    return coherence_lda

# Function to recommend content based on topics
def recommend_content(text, lda_model, dictionary):
    tokens = preprocess_text(text)
    bow = dictionary.doc2bow(tokens)
    topics = lda_model.get_document_topics(bow)
    recommended_topics = sorted(topics, key=lambda x: x[1], reverse=True)[:3]  # Get top 3 topics

    recommendations = []
    for topic_id, score in recommended_topics:
        topic_words = lda_model.show_topic(topic_id, topn=10)
        recommendations.append((topic_id, score, topic_words))

    return recommendations

# URL and keyword
url = 'https://thatware.io/'
keyword = 'Advanced Seo Service'

# Scrape website content
text = scrape_website(url)

# Preprocess and find LDA relevance
keyword_relevance, lda_model, dictionary, texts = find_lda_keyword_relevance(text, keyword)

# Calculate coherence score
coherence_score = calculate_coherence_score(lda_model, [texts], dictionary)

# Recommend content
recommendations = recommend_content(text, lda_model, dictionary)

# Output results
print(f'Keyword relevance in LDA topics for "{keyword}": {keyword_relevance}')
print(f'Coherence Score for LDA model: {coherence_score}')
print('Recommended Topics:')
for topic_id, score, topic_words in recommendations:
    print(f'Topic ID: {topic_id}, Score: {score}')
    print('Words:', topic_words)
```
